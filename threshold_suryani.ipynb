```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tkinter import Tk, filedialog

def pilih_file_gambar():
    root = Tk()
    root.withdraw()
    file_path = filedialog.askopenfilename()
    return file_path
```

```python
def tampilkan_gambar(title, image):
    plt.figure(figsize=(10, 10))
    if len(image.shape) == 2:  # Grayscale
        plt.imshow(image, cmap='gray')
    else:
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def terapkan_threshold(image, threshold_value):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, binary_image = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY)
    return gray_image, binary_image
```

```python
file_path = pilih_file_gambar()
if file_path:
    image = cv2.imread(file_path)
    
    # Tampilkan gambar asli
    tampilkan_gambar('Gambar Asli', image)
    
    # Terapkan thresholding
    threshold_value = 128  # Anda bisa mengganti nilai threshold sesuai kebutuhan
    gray_image, binary_image = terapkan_threshold(image, threshold_value)
    
    # Tampilkan gambar grayscale dan gambar biner
    tampilkan_gambar('Gambar Grayscale', gray_image)
    tampilkan_gambar('Gambar Biner (Thresholding)', binary_image)
else:
    print("Tidak ada file gambar yang dipilih.")
```
